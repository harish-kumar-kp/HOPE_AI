```python
# print 0 to 20 by using range
for item in range(0,20):
    print(item)
```

```python
# print range 10 to 20
for item in range(10,20):
    print(item, end =' ')
```

```python
# Print number of items in the list by using 'len'
myList = [10, 20, 14, 55, 43, 87, 76]
print(myList)
print("Number of items in the myList:\n", len(myList))
```

```python
word = "Artificial Intelligence"
print(word)
for letter in word:
    print(letter)    
```

```python
myList = ["-Your Name-","-Your Age-","-Your Profession-"]
for item in myList:
    print(item)
```

```python
# Print mixed datatype using Tuples
MyTuple = (1, 'Welcome', 2, 'Hope')
print(MyTuple)
```

```python
Tuple1 = (0, 1, 2, 3)
Tuple2 = ('python', 'HOPE')
Tuple3 = (Tuple1,Tuple2)
print(Tuple3)
```

```python
# print Odd Numbers in the list
MyTuple = (20,10,16,19,25,1,276,188)
print(MyTuple)
for number in MyTuple:
    if number % 2 == 1:
        print(number, "is odd")
        
```

```python
# print Even Numbers in the list
MyTuple = (20,10,16,19,25,1,276,188)
print(MyTuple)
for number in MyTuple:
    if number % 2 == 0:
        print(number, "is even")
```
