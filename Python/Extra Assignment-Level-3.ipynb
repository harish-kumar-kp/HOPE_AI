```python
# Check the password, using if and else
passWord = input("Enter the password:")
print("Enter the password:",passWord)
if passWord == "HOPE@123":
    print("Your password is correct")
else:
    print("Your password is wrong")
```

```python
# Catagory the people by their age like children, adult, citizen, senior citizen...
yourAge = int(input("age:"))
print("age:",yourAge)
if(yourAge < 18):
    print("Children")
elif(yourAge <= 35):
    print("Adult")
elif(yourAge <= 59):
    print("Citizen")
else:
    print("Senior Citizen")
```

```python
# Find whether given number is positive or negative
yourNumber = int(input("Enter any number:"))
print("Enter any number:",yourNumber)
if yourNumber > 0:
    print("No is positive")
elif yourNumber == 0:
    print("No is zero")
else:
    print("No is negative")
```

```python
# Check whether the given number is divisible by 5
myNumber = int(input("Enter a number to check:"))
print("Enter a number to check:",myNumber)
if myNumber % 5 == 0:
    print("No is divisible by 5")
else:
    print("No is not divisible by 5")
```
